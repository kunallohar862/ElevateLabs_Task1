```python
import pandas as pd
```

```python
data = pd.read_csv('/content/Index2018.csv')
data2 = data
```

```python
data.head()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['date'] = pd.to_datetime(data['date'], dayfirst = True)
data.info()
```

```python
data.set_index('date', inplace=True)
data.head()
```

```python
# Keeping only business days and filtering rest
df_crop = data.asfreq('B')
print(df_crop.shape)
df_crop.head()
```

```python
df_crop.isnull().sum()
```

# **Method 1 Using Foreward Fill**

```python
# foreward fill for missing values
# ffill for foreward fill
# bfill for backward fill
df_crop.fillna(method='ffill', inplace=True)
df_crop.isnull().sum()
```

# **Method 2 Using Mean**

```python
# prompt: fill missing values in all columns using mean

# Fill missing values with the mean of each column
df_crop.fillna(df_crop.mean(), inplace=True)
df_crop.isnull().sum()
df_crop
```

```python
data2.isnull().sum()
```

```python
spx_mean = data['spx'].mean()
dax_mean = data['dax'].mean()
ftse_mean = data['ftse'].mean()
nikkei_mean = data['nikkei'].mean()
```

```python
data2['spx'] = data2['spx'].fillna(spx_mean)
data2['dax'] = data2['dax'].fillna(dax_mean)
data2['ftse'] = data2['ftse'].fillna(ftse_mean)
data2['nikkei'] = data2['nikkei'].fillna(nikkei_mean)
```
